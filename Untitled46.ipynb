````python
import math
import numpy as _np
from typing import List, Tuple
from dataclasses import dataclass

try:
    import numba as _nb
except ImportError:   # graceful degradation
    class _nb:
        @staticmethod
        def njit(*args, **kwargs):
            def decorator(fn): return fn
            return decorator
        prange = range  # type: ignore

# GPU optional
try:
    import cupy as _cp
    _GPU_AVAILABLE = True
except ImportError:
    _GPU_AVAILABLE = False

# MPI optional
try:
    from mpi4py import MPI as _MPI
    _MPI_AVAILABLE = True
except ImportError:
    _MPI_AVAILABLE = False

# ─────────────────────────────────────────────────────────────────────────────
# 1. Sutra Engine
# ─────────────────────────────────────────────────────────────────────────────
@dataclass(frozen=True)
class Sutra:
    """Single Vedic sutra Δᵢ."""
    label: str
    delta: float

    def apply(self, x: float) -> float:
        """Apply multiplicative correction."""
        return x * (1.0 + self.delta)

    def inverse(self, x: float) -> float:
        """Remove correction."""
        return x / (1.0 + self.delta)


def build_29_sutras() -> List['Sutra']:
    """Generate the canonical ordered list of 29 sutras."""
    deltas = _np.linspace(1e-4, 2.9e-3, 29, dtype=_np.float64)
    labels = [f"S{i+1:02d}" for i in range(29)]
    return [Sutra(label=l, delta=d) for l, d in zip(labels, deltas)]


# ─────────────────────────────────────────────────────────────────────────────
# 2. Hypercube Geometry
# ─────────────────────────────────────────────────────────────────────────────
def kronecker_fabric_Q(d: int, chi: float = 1.0) -> _np.ndarray:
    """Return dense adjacency matrix of d-dimensional hypercube Q_d(χ)."""
    if d < 1:
        raise ValueError("Dimension d must be >= 1")
    n = 1 << d
    fabric = _np.zeros((n, n), dtype=_np.int8)
    for i in range(n):
        for j in range(d):
            fabric[i, i ^ (1 << j)] = 1
    _np.fill_diagonal(fabric, int(round(chi)))
    return fabric


@_nb.njit(parallel=True, fastmath=True)
def adjacency_fusion_P(fabric: _np.ndarray) -> _np.ndarray:
    """Compute boolean path-length-2 fusion P = A ⋅ A (OR-semiring)."""
    n = fabric.shape[0]
    out = _np.zeros_like(fabric)
    for i in _nb.prange(n):
        for k in range(n):
            if fabric[i, k]:
                for j in range(n):
                    if fabric[k, j]:
                        out[i, j] = 1
    return out


# ─────────────────────────────────────────────────────────────────────────────
# 3. Core Memory Engine
# ─────────────────────────────────────────────────────────────────────────────
class VedicHypercubeMemory:
    """Bidirectional bit ↔ geometry codec with sutra modulation."""

    def __init__(self, dimension: int, chi: float = 1.0):
        self.d = dimension
        self.chi = chi
        self.sutras: List[Sutra] = build_29_sutras()
        self.fabric_Q = kronecker_fabric_Q(dimension, chi)
        self.fabric_P = adjacency_fusion_P(self.fabric_Q)
        self.node_count = 1 << dimension

    # ─── Low-level helpers ─────────────────────────────────────────────────
    def _bits_from_bytes(self, data: bytes) -> _np.ndarray:
        return _np.unpackbits(_np.frombuffer(data, dtype=_np.uint8))

    def _bytes_from_bits(self, bits: _np.ndarray) -> bytes:
        if bits.size % 8:
            bits = _np.pad(bits, (0, 8 - bits.size % 8))
        return _np.packbits(bits.astype(_np.uint8)).tobytes()

    # ─── Encode / Decode ───────────────────────────────────────────────────
    def encode(self, data: bytes) -> List[int]:
        bits = self._bits_from_bytes(data)
        path = [0]
        cur = 0
        for bit in bits:
            hop_dim = int(bit) % self.d
            cur ^= 1 << hop_dim
            path.append(cur)
        return path

    def decode(self, path: List[int]) -> bytes:
        hops = _np.diff(_np.array(path, dtype=_np.int64))
        bits = _np.array([int(math.log2(abs(h))) & 1 for h in hops], dtype=_np.uint8)
        return self._bytes_from_bits(bits)

    # ─── Sutra-augmented write / read ──────────────────────────────────────
    def write(self, data: bytes, *, intensity: float = 1.0) -> Tuple[List[int], _np.ndarray]:
        path = self.encode(data)
        amps = _np.full(len(path), intensity, dtype=_np.float64)
        # Apply sutra modulation cyclically
        for idx, s in enumerate(self.sutras):
            amps[idx % amps.size] = s.apply(amps[idx % amps.size])
        return path, amps

    def read(self, path: List[int], amps: _np.ndarray) -> bytes:
        corr = amps.copy()
        for idx, s in enumerate(self.sutras):
            corr[idx % corr.size] = s.inverse(corr[idx % corr.size])
        return self.decode(path)

    # ─── Text message embedding / retrieval ────────────────────────────────
    def embed_message(self, message: str, *, amplitude_base: float = 1.0) -> Tuple[List[int], _np.ndarray]:
        """Embed arbitrary UTF-8 text as amplitude flips ±ϵ."""
        bytes_msg = message.encode("utf-8")
        bits = self._bits_from_bytes(bytes_msg)
        pad_len = (self.d - (bits.size % self.d)) % self.d
        bits = _np.pad(bits, (0, pad_len))  # align to hop groups

        # Map bits onto geometric hops
        path = [0]
        cur = 0
        signal = _np.full(bits.size + 1, amplitude_base, dtype=_np.float64)
        epsilon = 0.0073  # small, invertible offset
        for idx, bit in enumerate(bits):
            hop_dim = int(bit) % self.d
            cur ^= 1 << hop_dim
            path.append(cur)
            signal[idx + 1] += epsilon if bit else -epsilon
        # Sutra modulation
        for idx, s in enumerate(self.sutras):
            signal[idx % signal.size] = s.apply(signal[idx % signal.size])
        return path, signal

    def retrieve_message(self, path: List[int], amps: _np.ndarray) -> str:
        """Retrieve UTF-8 text embedded via embed_message."""
        corr = amps.copy()
        for idx, s in enumerate(self.sutras):
            corr[idx % corr.size] = s.inverse(corr[idx % corr.size])

        # Remove base amplitude and epsilon
        epsilon = 0.0073
        bits = _np.array([1 if abs(a - 1.0 - epsilon) < abs(a - 1.0 + epsilon) else 0 for a in corr[1:]], dtype=_np.uint8)
        raw_bytes = self._bytes_from_bits(bits)
        # Strip null padding
        raw_bytes = raw_bytes.rstrip(b"\x00")
        return raw_bytes.decode("utf-8", errors="ignore")

    # ─── Verification helpers ──────────────────────────────────────────────
    def verify_roundtrip(self, trials: int = 128):
        rng = _np.random.default_rng()
        for _ in range(trials):
            n = rng.integers(1, 256)
            dat = rng.integers(0, 256, size=n, dtype=_np.uint8).tobytes()
            assert self.read(*self.write(dat)) == dat

    def verify_message(self, trials: int = 64):
        rng = _np.random.default_rng()
        for _ in range(trials):
            msg_len = rng.integers(1, 128)
            text = ''.join(rng.choice(list("abcdefghijklmnopqrstuvwxyz"), msg_len))
            assert self.retrieve_message(*self.embed_message(text)) == text


# ─────────────────────────────────────────────────────────────────────────────
# 4. CLI Driver (Only used when script is run directly)
# ─────────────────────────────────────────────────────────────────────────────
def _cli():
    import argparse, sys, json
    from pathlib import Path
    p = argparse.ArgumentParser(description="VGME CLI – encode/decode & embed text")
    sub = p.add_subparsers(dest="cmd", required=True)
    # encode/decode
    ec = sub.add_parser("encode", help="Encode binary file to path+amps JSON")
    ec.add_argument("infile"), ec.add_argument("outfile")
    ec.add_argument("--dim", type=int, default=8)
    # decode
    dc = sub.add_parser("decode", help="Decode path+amps JSON to binary")
    dc.add_argument("infile"), dc.add_argument("outfile")
    # embed
    em = sub.add_parser("embed", help="Embed UTF-8 text; output JSON")
    em.add_argument("--dim", type=int, default=8)
    em.add_argument("message"), em.add_argument("outfile")
    # retrieve
    rt = sub.add_parser("retrieve", help="Retrieve text from JSON")
    rt.add_argument("infile")
    args = p.parse_args()

    vm = VedicHypercubeMemory(dimension=getattr(args, "dim", 8))
    if args.cmd == "encode":
        raw = Path(args.infile).read_bytes()
        path, amps = vm.write(raw)
        _save_json(args.outfile, path, amps)
    elif args.cmd == "decode":
        path, amps = _load_json(args.infile)
        Path(args.outfile).write_bytes(vm.read(path, amps))
    elif args.cmd == "embed":
        path, amps = vm.embed_message(args.message)
        _save_json(args.outfile, path, amps)
    elif args.cmd == "retrieve":
        path, amps = _load_json(args.infile)
        print(vm.retrieve_message(path, amps))
    else:
        p.error("Unknown command")


def _save_json(fname: str, path: List[int], amps: _np.ndarray):
    import json
    with open(fname, "w") as fp:
        json.dump({"path": path, "amps": amps.tolist()}, fp)


def _load_json(fname: str) -> Tuple[List[int], _np.ndarray]:
    import json
    from pathlib import Path
    j = json.loads(Path(fname).read_text())
    return j["path"], _np.array(j["amps"], dtype=_np.float64)


# ─────────────────────────────────────────────────────────────────────────────
# 5. Embedded LaTeX derivation
# ─────────────────────────────────────────────────────────────────────────────
LATEX_DERIVATION = r"""
\section{Proof of Geometry–Memory Bijectivity}
Let $\gamma = (\gamma_0,\gamma_1,\dots,\gamma_N)$ be the vertex sequence
returned by \texttt{encode}. Each hop satisfies
$\operatorname{Hamming}(\gamma_k,\gamma_{k+1})=1$,
mapping directly to a single bit $b_k = \log_2|\gamma_k\oplus\gamma_{k+1}|$.
Thus the inverse map $\gamma\mapsto b$ is uniquely defined.

Sutra modulation scales amplitudes by $(1+\Delta_i)$; since
$\Delta_i\in(0,0.005)\subset\mathbb{R}$ and never equals $-1$, multiplication
by $(1+\Delta_i)$ is bijective with inverse $(1+\Delta_i)^{-1}$.
Amplitude perturbation $\pm\epsilon$ encodes message bits.  Because
$\epsilon<\min_i\Delta_i$, corrections commute and inversion recovers the
unmodulated base allowing exact bit discrimination.

Therefore the composite map
$$
\texttt{embed}\;:\;\{0,1\}^{8m}\to Q_d\times\mathbb{R}^{N+1}
$$
is bijective, completing the proof. \qedhere
"""


# ─────────────────────────────────────────────────────────────────────────────
# 7. Optional GPU & MPI Acceleration
# ─────────────────────────────────────────────────────────────────────────────
if _GPU_AVAILABLE:
    # ---------------------------------------------------------------------
    # Cupy-based fast bit-manipulation kernels
    # ---------------------------------------------------------------------
    _cupy_encode_kernel = _cp.ElementwiseKernel(
        in_params="raw uint8 bits, uint32 d",
        out_params="raw uint32 path",
        operation=r"""
        const uint32 hop_dim = bits[i] & (d - 1);
        const uint32 mask = 1u << hop_dim;
        path[i + 1] = path[i] ^ mask;
        """,
        name="vgme_cu_encode",
        preamble="#include <stdint.h>"
    )

    def gpu_encode(data: bytes, d: int) -> _cp.ndarray:
        """
        Encode bytes → CUDA path (uint32 array).  Requires dimension d ≤ 32.
        """
        if d > 32:
            raise ValueError("GPU encode currently limited to d ≤ 32")
        bits = _cp.unpackbits(_cp.asarray(memoryview(data), dtype=_cp.uint8))
        path = _cp.zeros(bits.size + 1, dtype=_cp.uint32)
        _cupy_encode_kernel(bits, _cp.uint32(d), path)
        return path

    def gpu_decode(path: _cp.ndarray) -> bytes:
        """
        Decode CUDA path → bytes.  Inverse of gpu_encode.
        """
        hops = _cp.diff(path)
        bits = ((hops & -hops) != 0).astype(_cp.uint8)  # LSB test
        if bits.size % 8:
            bits = _cp.pad(bits, (0, 8 - bits.size % 8))
        return _cp.asnumpy(_cp.packbits(bits)).tobytes()


if _MPI_AVAILABLE:
    # ---------------------------------------------------------------------
    # MPI verification harness (lossless round-trip & message tests)
    # ---------------------------------------------------------------------
    def mpi_verify(dim: int = 8, trials: int = 1024):
        """
        Distribute random round-trip tests across MPI ranks; gather failures.
        """
        comm = _MPI.COMM_WORLD
        rank = comm.Get_rank()
        size = comm.Get_size()
        rng = _np.random.default_rng(seed=rank * 911)
        vm = VedicHypercubeMemory(dim)
        local_fail = 0
        for _ in range(trials // size + (rank < trials % size)):
            n = rng.integers(1, 256)
            blob = rng.integers(0, 256, size=n, dtype=_np.uint8).tobytes()
            if vm.read(*vm.write(blob)) != blob:
                local_fail += 1
        fails = comm.reduce(local_fail, op=_MPI.SUM, root=0)
        if rank == 0:
            print(f"MPI verify: {trials} trials, failures = {fails}")
            assert fails == 0, "Round-trip failures detected!"


# ─────────────────────────────────────────────────────────────────────────────
# 8. Export LaTeX paper utility
# ─────────────────────────────────────────────────────────────────────────────
def export_latex(outfile: str = "vgme_proof.tex"):
    """
    Dump `LATEX_DERIVATION` to disk.
    """
    from pathlib import Path
    Path(outfile).write_text(LATEX_DERIVATION, encoding="utf-8")
    print(f"LaTeX proof written → {outfile!s}")


# ─────────────────────────────────────────────────────────────────────────────
# 10. Advanced Analytics & Diagnostics
# ─────────────────────────────────────────────────────────────────────────────
def sutra_matrix() -> _np.ndarray:
    """
    Return 29×29 diagonal matrix Σ with entries (1+Δᵢ) to inspect eigen-profile.
    """
    s = build_29_sutras()
    return _np.diag([1.0 + st.delta for st in s])


def radial_geometry_entropy(path: List[int]) -> float:
    """
    Shannon entropy of radial hop distances from origin for a given path.
    Useful to quantify geometric dispersion of a memory trace.
    """
    hops = _np.array(path, dtype=_np.int64)
    radii = _np.array([bin(v).count("1") for v in hops])
    vals, counts = _np.unique(radii, return_counts=True)
    p = counts / counts.sum()
    return -_np.sum(p * _np.log2(p))


def geomviz_ascii(path: List[int], width: int = 80) -> str:
    """
    Simple ASCII sparkline visualising Hamming radius along the path.
    """
    radii = [bin(v).count("1") for v in path]
    max_r = max(radii)
    scale = width / (len(radii) - 1)
    canvas = [" "] * width
    for idx, r in enumerate(radii):
        pos = int(idx * scale)
        canvas[pos] = "." if r == 0 else "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"[r % 36]
    return "".join(canvas)


# ─────────────────────────────────────────────────────────────────────────────
# 11. Property-based Testing (optional, Hypothesis)
# ─────────────────────────────────────────────────────────────────────────────
try:
    from hypothesis import given, strategies as st

    @given(st.binary(min_size=0, max_size=256))
    def _hypothesis_roundtrip(blob: bytes):
        vm = VedicHypercubeMemory(8)
        assert vm.read(*vm.write(blob)) == blob

    @given(st.text(min_size=0, max_size=128))
    def _hypothesis_message(msg: str):
        vm = VedicHypercubeMemory(8)
        assert vm.retrieve_message(*vm.embed_message(msg)) == msg

except ImportError:
    # Hypothesis not available; skip property tests.
    pass


# ─────────────────────────────────────────────────────────────────────────────
# 12. Extended Benchmarks
# ─────────────────────────────────────────────────────────────────────────────
def bench_encode_decode(dim: int = 12, n_bytes: int = 1_000_000, loops: int = 5):
    """
    Micro-benchmark pure-CPU encode/decode throughput.
    """
    import time, sys
    rng = _np.random.default_rng(123)
    blob = rng.integers(0, 256, size=n_bytes, dtype=_np.uint8).tobytes()
    vm = VedicHypercubeMemory(dim)
    t0 = time.perf_counter()
    for _ in range(loops):
        path = vm.encode(blob)
    enc_t = time.perf_counter() - t0
    t0 = time.perf_counter()
    for _ in range(loops):
        vm.decode(path)
    dec_t = time.perf_counter() - t0
    print(f"Encode: {n_bytes*loops/enc_t/1e6:.2f} MB/s | "
          f"Decode: {n_bytes*loops/dec_t/1e6:.2f} MB/s")


# ─────────────────────────────────────────────────────────────────────────────
# 13. Self-test entry
# ─────────────────────────────────────────────────────────────────────────────
def run_selftest():
    """Run basic self-tests and verification."""
    print("[*] Verifying binary round-trip …", end=" ")
    VedicHypercubeMemory(8).verify_roundtrip(256)
    print("OK")
    print("[*] Verifying message embedding …", end=" ")
    VedicHypercubeMemory(8).verify_message(128)
    print("OK")
    print("[*] Entropy demo:")
    vm = VedicHypercubeMemory(6)
    p, a = vm.write(b"entropy")
    print("    ASCII:", geomviz_ascii(p))
    print(f"    H_r = {radial_geometry_entropy(p):.4f} bits")
    return "All tests passed successfully!"


# ─────────────────────────────────────────────────────────────────────────────
# 14. Demo functions for Jupyter/Colab
# ─────────────────────────────────────────────────────────────────────────────
def demo_string_embedding(text: str = "Hello VGME world!", dimension: int = 8):
    """
    Demonstrate text embedding with visualization.
    Returns path, amplitudes, and visualization.
    """
    vm = VedicHypercubeMemory(dimension)
    path, amps = vm.embed_message(text)
    viz = geomviz_ascii(path)
    entropy = radial_geometry_entropy(path)

    print(f"Input text: {text!r}")
    print(f"Path length: {len(path)}")
    print(f"Geometry visualization:")
    print(viz)
    print(f"Radial entropy: {entropy:.4f} bits")

    # Verify round-trip
    recovered = vm.retrieve_message(path, amps)
    print(f"Recovered text: {recovered!r}")
    assert recovered == text, "Round-trip error!"

    return path, amps, viz


def demo_binary_encoding(data: bytes = None, dimension: int = 8):
    """
    Demonstrate binary data encoding with visualization.
    If no data provided, uses a sample string.
    """
    vm = VedicHypercubeMemory(dimension)
    if data is None:
        data = b"This is a binary test for VGME" * 3

    path, amps = vm.write(data)
    viz = geomviz_ascii(path)
    entropy = radial_geometry_entropy(path)

    print(f"Input data size: {len(data)} bytes")
    print(f"Path length: {len(path)}")
    print(f"Geometry visualization:")
    print(viz)
    print(f"Radial entropy: {entropy:.4f} bits")

    # Verify round-trip
    recovered = vm.read(path, amps)
    assert recovered == data, "Round-trip error!"
    print("Round-trip verification: OK")

    return path, amps, viz


# ─────────────────────────────────────────────────────────────────────────────
# Main Guard
# ─────────────────────────────────────────────────────────────────────────────
if __name__ == "__main__":  # pragma: no cover
    import sys
    if len(sys.argv) > 1:
        if sys.argv[1] == "selftest":
            run_selftest()
            sys.exit(0)
        elif sys.argv[1] == "--gpu-bench":
            if not _GPU_AVAILABLE:
                print("CuPy not available – GPU bench skipped")
                sys.exit(1)
            vm = VedicHypercubeMemory(16)        # d ≤ 32 for GPU path
            raw = b"The quick brown fox jumps over the lazy dog" * 1024
            cp_path = gpu_encode(raw, vm.d)
            assert gpu_decode(cp_path)[:len(raw)] == raw
            print(f"GPU encode/decode OK – path length {cp_path.size}")
            sys.exit(0)
        elif sys.argv[1] == "--mpi-verify":
            if not _MPI_AVAILABLE:
                print("mpi4py not available – MPI verify skipped")
                sys.exit(1)
            mpi_verify()
            sys.exit(0)
        else:
            # Regular CLI mode
            _cli()
    else:
        # No args, print help
        print("VGME - Vedic-Geometric Memory Engine")
        print("Run with 'selftest', '--gpu-bench', or '--mpi-verify' for tests")
        print("Or import the module and use the VedicHypercubeMemory class directly")
````
